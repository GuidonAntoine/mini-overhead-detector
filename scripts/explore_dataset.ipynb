```python
# Path to data
racine_data = Path("../data/interim/nwpu-vhr-10/dataset")
gt_dir = racine_data / "ground_truth"
positive_dir = racine_data / "positive_image_set"
negative_dir = racine_data / "negative_image_set"
```

```python
CLASS_MAP = {
    1: "airplane",
    2: "ship",
    3: "storage tank",
    4: "baseball diamond",
    5: "tennis court",
    6: "basketball court",
    7: "ground track field",
    8: "harbor",
    9: "bridge",
    10: "vehicle",
}
```

```python
positive_images = sorted(positive_dir.glob("*.jpg"))
negative_images = sorted(negative_dir.glob("*.jpg"))
annotation_files = sorted(gt_dir.glob("*.txt"))

```

```python
# Compter le dataset 

print(f"Positive images: {len(positive_images)}")
print(f"Negative images: {len(negative_images)}")
print(f"Annotation files: {len(annotation_files)}")
print(f"Total images: {len(positive_images) + len(negative_images)}")
```

```python
# Check the dataset quality
positive_ids = {path.stem for path in positive_images}
annotation_ids = {path.stem for path in annotation_files}

print("Positive IDs:", len(positive_ids))
print("Annotation IDs:", len(annotation_ids))
print("Missing annotations:", len(positive_ids - annotation_ids))
print("Annotations without image:", len(annotation_ids - positive_ids))
```

```python
def parse_annotation_file(file_path):
    annotations = []

    with file_path.open("r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            line = line.replace("(", "").replace(")", "")
            parts = line.split(",")

            x1, y1, x2, y2, class_id = map(int, parts)

            annotations.append({
                "Annotations": CLASS_MAP[class_id],
                "bbox": [x1, y1, x2, y2],
            })

    return annotations
```

```python
samples = []
for gt_files in annotation_files:
    annotations = parse_annotation_file(gt_files)
    image_name = gt_files.stem+".jpg"
    image_path=positive_dir/image_name
    
    sample = {
        "image_path" : image_path,
        "annotations" : annotations
    }
    
    samples.append(sample)
```
